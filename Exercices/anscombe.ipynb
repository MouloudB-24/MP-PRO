# 1. Analysez la relation entre x et y.

```python
sns.scatterplot(data=df, x='x', y='y', hue='dataset', alpha=0.7)
```

- Il semblerait que lorsque x augmente,  la moyenne de y augmente également, mais manière non linéaire.

# 2. Y a-t-il une meilleure manière d’analyser les choses ? Adaptez votre graphique

```python
sns.relplot(data=df, x='x', y='y', col='dataset')
```

- Il semblerait qu'il y a 4 goupes de données
   - Le groupe I suite une échelle léneaire avec du bruit
   - Le groupe II suite une échelle quadratique
   - Le groupe III suite une échelle linéaire, avec un outlier
   - Le goupe IV suite une colonne, avec un outlier

# 3. Analysez maintenant les statistiques de chaque "paquet". Que constatez-vous ?

```python
df.groupby('dataset').describe()
```

- La moyenne et l'écart-type des variable x et y sont idéntique pour tous les paquets du jeu de données.

# 4. Quels sont les 2 enseignements que vous pouvez tirer de cette analyse ?

- Dans un même jeu de données, des sous-groupes plus significatifs peuvent se cacher
- Sous une seule variable se cache souvent plusieurs signaux !
- Avec les mêmes statiques, il est possible de créer plusieurs jeu de données différents.
- Les statistiques sont parfois trompeuses
