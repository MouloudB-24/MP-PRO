## 3. Analysez la variable "cylinders"
 - Plus de 50% de modèles de voitures ont 4 cylindres. Est-ce la meilleure configuration ?
 - Plus de 99% de modèles de voitures ont un nombre de cylindres pair (4, 6 ou 8). Les modèles avec 3 et 5 cylindres sont-ils moins performants ?

```python
df['cylinders'].value_counts(normalize=True).round(2).sort_index()
```

```python
df['cylinders'].value_counts().sort_index().plot.bar()
plt.title("Le nombre d'apparitions des différents modèles de voitures en fonction du nombre \nde cylindres qu'ils contiennent" )
plt.show()
```

## 4. Analysez la variable "origin"

```python
df['origin'].value_counts(normalize=True).round(2)
```

```python
sns.countplot(data=df, x='origin')
plt.show()
```

## 5. Observez l'évolution de la consommation des voitures (mpg) par rapport au nombre de chevaux (horsepower) 
 - La distance parcourue en mile par les véhicules pour un gallon diminue à mesure que la puissance des véhicules augmente
 - Il semble que les véhicules les plus récents, ceux datant de 80 à 82, sont en général les plus performants en termes de distance parcounue par gallon

```python
sns.scatterplot(data=df, x='horsepower', y='mpg', hue='model_year', s=100, alpha=0.8)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(data=df, x='horsepower', y='mpg')
plt.xticks(rotation=90)
plt.show()
```

## Observez l'évolution du poids moyen des voitures américaines de l'année 70 jusqu'à l'année 82.

- Le poids moyen des véhicules aux Etats-Unis a diminué au fil du temps, de 1970 à 1982. Le poids moyen des véhicules en Europe et au Japan est resté relativement stable sur cette période.



```python
usa = pd.DataFrame(df.query("origin=='usa'").groupby('model_year')['weight'].mean())
europe = pd.DataFrame(df.query("origin=='europe'").groupby('model_year')['weight'].mean())
japan = pd.DataFrame(df.query("origin=='japan'").groupby('model_year')['weight'].mean())
```

```python
#fig, ax = plt.subplots(3, 1, sharex=True, figsize=(6, 12))
plt.figure(figsize=(8, 6))
plt.title('La distribution du poids moyen des véhicules aux États-Unis, Europe et Japan, de 1970 à 1980')
sns.scatterplot(data=usa, x='model_year', y='weight', label='USA')
sns.scatterplot(data=europe, x='model_year', y='weight', label='Europe')
sns.scatterplot(data=japan, x='model_year', y='weight', label='Japan')
plt.show()
```

## 7. Analysez la relation entre le nombre de cylindres et l'année du modèle.

- La fabrication des véhicules de 3 et 5 cylindres est moins fréquente que celle de 4, 6 et 8 cylindres. Suelement 4 véhuciles de 3 cylindres, et 3 véhicules de 5 cylindres ont été fabriqués entre 1970 et 1982.
- La production des véhicules à 4, 6 et 8 cylindres est resté relativement stable au fil du temps, de 1970 à 1980. En revanche, celle des modèles à 3 et 5 cylindres a été plus pontuelle durant cette période
- La production moyenne des véhicules à 4 cylindres est doublé à partir de 1980

```python
c = df['cylinders']
m = df['model_year']
pd.crosstab(m, [c])
```

## 8. Analysez la relation entre le nombre de cylindres et l'origine.

- Les États-Unis, Japon et Europe ont fabriqué des véhicules de 4 et 6 cylindres entre 1970 et 1982
- Les véhicules de 3 cylindres ont été uniquement fabriqués au Japon durant cette période, mais en faible quantité
- Ceux à 5 cylindres ont été uniquement produit en Europe,  également en faible quantité
- Enfin les véhicules à 8 cylindres ont été exclusivement fabriqués aux États-Unis, en grande quantité

```python
o = df['origin']
c = df['cylinders']

pd.crosstab(o, [c])
```

## 9. Quelle est l'accélération moyenne des voitures des différents pays ?
 - Europe et Japan : 16 m/s2
 - USA : 15 m/s2

```python
df.groupby('origin')['acceleration'].mean().round(2)
```

## 10. Quelle est-elle pour les modèles de l'année 80 ?
 - Europe : 18 m/s2
 - Japan : 16 m/s2
 - USA : 17 m/s2

```python
df.query("model_year == 80").groupby('origin')['acceleration'].mean().round(2)
```

## 11. Au final Il semblerait qu'une tendance se dégage à travers le temps...
- Le poids moyen des véhicules aux Etats-Unis a diminué au fil du temps, de 1970 à 1982. Le poids moyen des véhicules en Europe et au Japan est resté relativement stable sur cette période.
- L'accélération moyenne des véhicules a augmenté à partir des années 1980, en Europe et aux États-Unis. En revanche, celle du Japon est resté relativement stable
- La puissance des véhicules diminue au fil du temps, de 1970 à 1982, aux États-Unis, en Europe et au Japon
- La distance parcourue pour un gallon des véhicules augmente au fil du temps, de 1970 à 1982, aux États-Unis, en Europe et au Japon
-  Le volume des moteurs a diminu au fil du temps, pour les véhicules produits aux États-Unis. En revanche, celui des véhicules produits en Europe et au Japon est resté relativement stable 

```python
df_group = pd.DataFrame(df.groupby(['model_year', 'origin'])[['horsepower', 'mpg', 'displacement']].mean().reset_index())
```

```python
fig, ax = plt.subplots(3, 3, sharex=True, figsize=(12, 8))
sns.scatterplot(data=df_group.query("origin=='usa'"), x='model_year', y='horsepower', ax=ax[0, 0])
sns.scatterplot(data=df_group.query("origin=='usa'"), x='model_year', y='mpg', ax=ax[1, 0])
sns.scatterplot(data=df_group.query("origin=='usa'"), x='model_year', y='displacement', ax=ax[2, 0])
ax[0, 0].set_title("USA", fontweight='bold')

sns.scatterplot(data=df_group.query("origin=='europe'"), x='model_year', y='horsepower', ax=ax[0, 1])
sns.scatterplot(data=df_group.query("origin=='europe'"), x='model_year', y='mpg', ax=ax[1, 1])
sns.scatterplot(data=df_group.query("origin=='europe'"), x='model_year', y='displacement', ax=ax[2, 1])
ax[0, 1].set_title("Europe", fontweight='bold')

sns.scatterplot(data=df_group.query("origin=='japan'"), x='model_year', y='horsepower', ax=ax[0, 2])
sns.scatterplot(data=df_group.query("origin=='japan'"), x='model_year', y='mpg', ax=ax[1, 2])
sns.scatterplot(data=df_group.query("origin=='japan'"), x='model_year', y='displacement', ax=ax[2, 2])
ax[0, 2].set_title("Japan", fontweight='bold')

plt.tight_layout()
plt.show()
```
